Combine 12 months data frame into one data frame which name is combined_data

```r
# Combine the twelve data frames into one using rbind
combined_data <- rbind(
  january_2021, february_2021, march_2021, april_2021, 
  may_2021, june_2021, july_2021, augest_2021, 
  september_2021, october_2021, november_2021, december_2021
)

# View the combined data frame
print(combined_data)
```

find out the ride length using ended_at minus started_at

```r
combined_data$ride_length= difftime(combined_data$ended_at, combined_data$started_at, units='mins')
combined_data
```

```r
library(lubridate)

# Convert started_at to a proper date-time format
combined_data$started_at <- ymd_hms(combined_data$started_at)

# Now, you can calculate the day of the week
combined_data$day_of_week <- wday(combined_data$started_at)

# Print the first few rows of the combined_data data frame
head(combined_data)
```

Count of bike types and difference between member vs casual 

```r



bike_type_counts <- table(combined_data$rideable_type	)

# Count the occurrences of each member type
member_casual_counts <- table(combined_data$member_casual)



# Create a bar plot for bike types
if (length(bike_type_counts) > 0) {
  barplot(bike_type_counts, main = "Count of Bike Types", xlab = "Bike Type", ylab = "Count", col = "blue")
} else {
  cat("No valid data for bike types.")
}

# Create a bar plot for member vs. casual
if (length(member_casual_counts) > 0) {
  barplot(member_casual_counts, main = "Member vs. Casual", xlab = "Member Type", ylab = "Count", col = "green")
} else {
  cat("No valid data for member vs. casual.")
}
```

Number of Bike Rides Per Week

```r
# Assuming 'combined_data' is your data frame

# Calculate the number of bike rides per day_of_week and member_casual
rides_per_week <- aggregate(ride_id ~ day_of_week + member_casual, data = combined_data, FUN = length)

# Create a bar plot using base R
barplot(rides_per_week$ride_id, beside = TRUE, col = c("blue", "green"),
        main = "Number of Bike Rides Per Week", xlab = "Day of Week", ylab = "Number of Rides",
        legend.text = unique(rides_per_week$member_casual), args.legend = list(x = "topright", bty = "n"),
        names.arg = rides_per_week$day_of_week)
```

The most popular start station for casual riders

```r
# Assuming 'old_combine_data' is your data frame
# Subset data for casual riders
casual_data <- old_combine_data[old_combine_data$member_casual == "casual", ]

# Count occurrences of start station names
start_station_counts <- table(casual_data$start_station_name)

# Find the most popular start station
most_popular_start_station <- names(start_station_counts)[which.max(start_station_counts)]
print(most_popular_start_station)

# Print the most popular start station
cat("The most popular start station for casual riders is:", most_popular_start_station, "\n")
```
